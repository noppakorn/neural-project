```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv("bangkok_traffy.csv")
```

```python
df = df.loc[:, ["comment", "type", "timestamp"]]
```

```python
df
```

```python
df.loc[:, "timestamp"] = pd.to_datetime(df.loc[:, "timestamp"])

start_date = pd.to_datetime("2023-01-01 00:00:00.000000+0000",)
df = df[df["timestamp"] >= start_date]

df = df.drop("timestamp", axis=1)
df = df.reset_index(drop=True)
```

```python
df
```

```python
df.to_csv("traffy-data-2023.csv", index=False)
```
